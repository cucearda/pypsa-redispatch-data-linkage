```python
import pandas as pd                                                                                                                                                                                                
```

```python
# full raw data
rd = pd.read_csv("Redispatch_Daten.csv", sep=";", encoding="utf-8-sig")                                                                                                                                            
rd.columns = rd.columns.str.strip()                                                                                                                                                                                
rd["BETROFFENE_ANLAGE"] = rd["BETROFFENE_ANLAGE"].str.strip()
                                                                                                                                                                                                                    
# lookup                                      
lk = pd.read_csv("plant_lookup_total_v1.csv", encoding="utf-8")    
```

```python
# join on plant name → bring in source column                                                                                                                                                                      
merged = rd.merge(                                                                                                                                                                                                 
    lk[["redispatch_name", "source"]],                                                                                                                                                                             
    left_on="BETROFFENE_ANLAGE",              
    right_on="redispatch_name",                                                                                                                                                                                    
    how="left",
)                                                                                                                                                                                                                  
merged["source"] = merged["source"].fillna("untouched")

# count raw redispatch rows per category                                                                                                                                                                           
counts = merged["source"].value_counts()
total  = len(merged)                                                                                                                                                                                               
                                            
print(counts)                                                                                                                                                                                                      
print("\nCoverage %:")
print((counts / total * 100).round(1))   
```
